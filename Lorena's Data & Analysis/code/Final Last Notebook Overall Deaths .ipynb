```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import json
from scipy.stats import linregress
```

```python
gender_rate="Total Overall Death Rate by Gender.csv"
gen_read=pd.read_csv(gender_rate, low_memory=False)
gen_read
```

```python
# Pandas can also plot multiple columns if the DataFrame includes them
multi_plot = gen_read.plot(kind="bar", figsize=(20,5))

# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(gen_read["Gender"], rotation=90)

plt.title('Male & Female: Overall Death Rate by All Causes, COVID & All Causes Except COVID')
plt.xlabel('Genders')
plt.ylabel('Death Rate')
plt.tight_layout()
plt.savefig("Male & Female: Overall Death Rate by All Causes, COVID & All Causes Except COVID")
plt.show()
```

```python
us_rate="US Overall Deaths & Death Rates.csv"
us_read=pd.read_csv(us_rate, low_memory=False)
us_read
```

```python
new_read=us_read.dropna()
new_read
```

```python
df2 = new_read[["Country","Covid Death Rate (Est 2019 Pop)","All Deaths Rate (Est 2019 Pop)", "All Deaths Except Covid Death Rate (Est 2019 Pop)"]]
# df3 = new_read[["Country","Covid Death Rate (Est 2020 Pop)","All Deaths Rate (Est 2020 Pop)", "All Deaths Except Covid Death Rate (Est 2020 Pop)"]]
df4 = new_read[["Country","Covid Deaths","Deaths from All Causes", "Deaths from All Causes Except Covid"]]
```

```python
# Pandas can also plot multiple columns if the DataFrame includes them
multi_plot = df2.plot(kind="bar", figsize=(20,5))

# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(new_read["Country"], rotation=90)

plt.title('Overall Death Rate by All Causes, COVID & All Causes Except COVID (Est 2019 Pop)')
plt.xlabel('Country')
plt.ylabel('Death Rate')
plt.tight_layout()
# plt.savefig("Male & Female: Overall Death Rate by All Causes, COVID & All Causes Except COVID")
plt.show()
```
